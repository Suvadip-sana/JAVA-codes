```python
im=cv2.imread("lena.tiff")
imblue=im[:,:,0]
imgreen=im[:,:,1]
imred=im[:,:,2]
im=cv2.cvtColor(im,cv2.COLOR_BGR2RGB)
plt.subplot(221)
plt.title("Original")
plt.imshow(im)
plt.subplot(222)
plt.title("Blue channel")
plt.imshow(imblue,cmap='gray')
plt.subplot(223)
plt.title("Green channel")
plt.imshow(imgreen,cmap='gray')
plt.subplot(224)
plt.title("Red channel")
plt.imshow(imred,cmap='gray')
plt.suptitle("Channel Seperation")
plt.show()
```

```python
im1=cv2.imread("lena.tiff")
im2=cv2.imread("splash.tiff")
im1=cv2.cvtColor(im1,cv2.COLOR_BGR2RGB)
im2=cv2.cvtColor(im2,cv2.COLOR_BGR2RGB)
```

```python
im1part=im1[200:400,200:400].copy()
im1[200:400,200:400]=im2[200:400,200:400]
im2[200:400,200:400]=im1part
```

```python
plt.subplot(121)
plt.title("Lena")
plt.imshow(im1)
plt.subplot(122)
plt.title("Splash")
plt.imshow(im2)
plt.show()
```
